```python
#download dataset
!wget https://d37ci6vzurychx.cloudfront.net/trip-data/green_tripdata_2025-11.parquet
```

```python
import pandas as pd
df = pd.read_parquet("green_tripdata_2025-11.parquet")
```

```python
#checking the data sample
df.head(2)
```

```python
# filter for the distance below 100
df_2 = df[df["trip_distance"] <= 100]
```

```python
df_2
```

```python
#Sort the trip distance with descending order
df_2[["lpep_pickup_datetime","trip_distance"]].sort_values(by='trip_distance', ascending=False).head(3)
```

Longest trip below 100 miles is 88.03 miles, which has pickup date at 2025-11-14, <br>
so the answer is <b> 2025-11-14 (a) </b>

For the passengers picked up in the zone named "East Harlem North" in November 2025, which was the drop off zone that had the largest tip? (1 point) <br>

a. JFK Airport <br>
b. Yorkville West <br>
c. East Harlem North <br>
d. LaGuardia Airport

```python
#download dataset green trip data
!wget https://d37ci6vzurychx.cloudfront.net/trip-data/green_tripdata_2025-11.parquet
```

```python
#download dataset zone data
!wget https://github.com/DataTalksClub/nyc-tlc-data/releases/download/misc/taxi_zone_lookup.csv
```

```python
import pandas as pd
trip_data = pd.read_parquet("green_tripdata_2025-11.parquet")
zone_data = pd.read_csv("taxi_zone_lookup.csv")
```

```python
#checking the data sample
trip_data.head(2)
```

```python
#checking the data sample
zone_data.head(2)
```

```python
#Identify the Dropout Zone by merging zone_data with trip data
df_dropout_zone = pd.merge(trip_data,
    zone_data,
    left_on="DOLocationID",
    right_on="LocationID",
    how="left"
)

df_dropout_zone.head(3)
```

```python
#Identify the Pickup Zone by merging zone_data with df_dropout_zone
df_pickup_zone = pd.merge(df_dropout_zone,
    zone_data,
    left_on="PULocationID",
    right_on="LocationID",
    how="left"
)

df_pickup_zone.head(3)
```

```python
#rename the column
df_2 = df_pickup_zone.rename(columns={"Zone_x": "zone_drop", "Zone_y": "zone_pickup"})
```

```python
#use groupby to calculate tip amount per zone drop
df_2.groupby("zone_drop").agg(total=("tip_amount", "sum")).sort_values("total", ascending=False)
```

The drop off zone that had the largest tip is <b> La Guardia Airport (d) </b>, which has value of 7254.51

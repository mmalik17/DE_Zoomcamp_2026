# QUESTION 3
For the trips in November 2025, how many trips had a trip_distance of less than or equal to 1 mile? <br>
a. 7,853 <br>
b. 8,007 <br>
c. 8,254 <br>
d. 8,421 <br>


```python
#download dataset
!wget https://d37ci6vzurychx.cloudfront.net/trip-data/green_tripdata_2025-11.parquet
```

```python
import pandas as pd
df = pd.read_parquet("green_tripdata_2025-11.parquet")
```

```python
#checking the data sample
df.head(2)
```

```python
print(df.dtypes)
```

```python
# add column for date filter
df['filter_date'] = (df["lpep_pickup_datetime"] >= "2025-11-01") & (df["lpep_pickup_datetime"] < "2025-12-01")

# add column for distance filter
df['filter_distance'] = df["trip_distance"] <= 1
```

```python
#checking whether the column is added or not
df.columns
```

```python
#check the total data that matches filter date and filter distance
len(df[(df['filter_date']==True) & (df['filter_distance']==True)])
```

the answer is 8007 (b)

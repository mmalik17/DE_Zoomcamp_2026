# QUESTION 5

Which was the pickup zone with the largest total_amount (sum of all trips) on November 18th, 2025? (1 point) <br>

a. East Harlem North <br>
b. East Harlem South <br>
c. Morningside Heights <br>
d. Forest Hills <br>


```python
#download dataset green trip data
!wget https://d37ci6vzurychx.cloudfront.net/trip-data/green_tripdata_2025-11.parquet
```

```python
#download dataset zone data
!wget https://github.com/DataTalksClub/nyc-tlc-data/releases/download/misc/taxi_zone_lookup.csv
```

```python
import pandas as pd
trip_data = pd.read_parquet("green_tripdata_2025-11.parquet")
zone_data = pd.read_csv("taxi_zone_lookup.csv")
```

```python
#checking the data sample
trip_data.head(2)
```

```python
#checking the data sample
zone_data.head(2)
```

```python
#JOIN two table
df_trip_zone = trip_data.merge(
    zone_data,
    left_on="PULocationID",
    right_on="LocationID",
    how="left"
)

df_trip_zone.head(3)
```

```python
#Filter only trip on 2025-11-18
df_day = df_trip_zone[
    df_trip_zone["lpep_pickup_datetime"].dt.date == pd.to_datetime("2025-11-18").date()
]

#Grouping the trip on 2025-11-18 by Zone
zone_group = df_day.groupby("Zone")["total_amount"].sum().reset_index()
```

```python
#Sorting the total_amount
zone_group.sort_values(by = 'total_amount', ascending = False)
```

The greatest total_amount is 9281.92, which is located at East Harlem North, <br>
so the answer is <b> East Harlem North (a) </b>
